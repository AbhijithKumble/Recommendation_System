```python
def getTopFiveProducerFromList(producer: str):
    if not isinstance(producer, str) or not producer.strip():
        return ""
    
    top_five = producer.split(',')[:5]

    # Remove all spaces and lowercase each producer name
    top_five_clean = [p.replace(" ", "").lower() for p in top_five]

    return ' '.join(top_five_clean)
```

```python
df_final.loc[:,'producer'] = df_final['producer'].apply(getTopFiveProducerFromList)
```

```python
df_final['producer'] = df_final['producer'].fillna('')
```

```python
df_final['rating'] =  df_final['rating'].fillna('R - 17+ (violence & profanity)')
```

```python
df_final = df_final.dropna()
```

```python
def cleanRating(rating: str):
    if not isinstance(rating, str) or not rating.strip():
        return ""
    
    return rating.replace(" ", "").lower()
```

```python
df_final.loc[:,'rating'] = df_final.rating.apply(cleanRating)
```

```python
def cleanStatus(status: str):
    if not isinstance(status, str) or not status.strip():
        return ""
    
    return status.replace(" ", "").lower()
```

```python
df_final.loc[:,'status']=df_final.status.apply(cleanStatus)
```

```python
def getTopThreeStudio(studios: str):
    if not isinstance(studios, str) or not studios.strip():
        return ""
    
    top_three = studios.split(',')[:3]
    
    # Remove all spaces and lowercase each studio name
    top_three_clean = [s.replace(" ", "").lower() for s in top_three]
    
    return ' '.join(top_three_clean)
```

```python
 df_final.loc[:, 'studio'] = df_final.loc[:,'studio'].apply(getTopThreeStudio)
```

```python
def splitGenre(genre: str):
    if not isinstance(genre, str) or not genre.strip():
        return ""
    
    genre_list = genre.split(',')
    
    # Remove all spaces and lowercase each genre
    genre_clean = [g.replace(" ", "").lower() for g in genre_list]
    
    return ' '.join(genre_clean)
```

```python
df_final.loc[:, 'genre'] = df_final.loc[:, 'genre'].apply(splitGenre)
```

```python
## convert all other cols to tags
exclude_cols = ['anime_id', 'title']  # columns to exclude

def flatten_cell(cell):
    if isinstance(cell, list):
        return ' '.join(s.strip() for s in cell if isinstance(s, str))
    return str(cell).strip()

df_final.loc[:, 'tags'] = df_final.apply(
    lambda row: ' '.join(flatten_cell(row[col]) for col in df_final.columns if col not in exclude_cols),
    axis=1
)
```

```python
df_final = df_final[['anime_id','title','tags']]
```

```python
df_final.loc[:,'tags'] = df_final['tags'].str.lower()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df_final.loc[:,'tags'] = df_final.loc[:, 'tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=3000, stop_words='english')
```

```python
vectors = cv.fit_transform(df_final['tags']).toarray()
```

```python
list(cv.get_feature_names_out())[2000:] ## too many numbers 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
result = cosine_similarity(vectors)
```

```python
similarity_matrix = result.copy()
```

```python
import difflib

def recommend(movie):
    '''
    Return 5 most similar movies
    '''
    if movie not in df_final['title'].values:
        close_matches = difflib.get_close_matches(movie, df_final['title'], n=1, cutoff=0.6)
        if close_matches:
            print(f"Did you mean: '{close_matches[0]}'?")
        else:
            print("No movie found.")
        return

    movie_index = df_final[df_final['title'] == movie].index[0]
    distances = similarity_matrix[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(i[0])
```

```python
import pickle

# Load your movie dataframe and similarity matrix
with open("../models/animes.pkl", "rb") as f:
    movies = pickle.load(f)

with open("../models/similarity_matrix.pkl", "rb") as f:
    similarity_matrix = pickle.load(f)

# Build the top-5 similar dict
top_5_similar = {}

for i in range(len(movies)):
    sim_scores = list(enumerate(similarity_matrix[i]))
    # Sort by similarity score, descending; skip the first (it's the movie itself)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]
    
    movie_title = movies.iloc[i]['title']
    movie_id = movies.iloc[i]['anime_id']

    top_5_similar[movie_title] = [
        {
            'title': movies.iloc[j]['title'],
            'anime_id': movies.iloc[j]['anime_id']
        }
        for j, _ in sim_scores
    ]

# Save the compact dictionary
with open("../models/top_5_similar.pkl", "wb") as f:
    pickle.dump(top_5_similar, f)

print("✅ Saved top_5_similar.pkl")
```
